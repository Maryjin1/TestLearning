```python
import pandas as pd 
import community
import networkx as nx
```

```python
def sort_shop(row_df):
	print "shop ranking..."
	names = ["cardaccptrid", "userid", "outmoney"]
	df = row_df[names]

	trade_df = df[['cardaccptrid', 'userid']]
	unique_df = trade_df[trade_df.duplicated()].groupby("cardaccptrid").agg({"userid":"count"}).reset_index()
	unique_df.rename(columns={'userid':'userid_cnt'}, inplace=True)
	new_df = df.groupby("cardaccptrid").agg({"userid":'count', "outmoney": 'sum'}).reset_index()
	new_df.rename(columns={"userid":"trade_cnt"}, inplace=True)

	merge_df = pd.merge(unique_df, new_df, on=['cardaccptrid'] )
	merge_df['trade_avg'] = merge_df["trade_cnt"]*1.0/merge_df["userid_cnt"]
	merge_df['money_avg'] = merge_df["outmoney"]/ merge_df["userid_cnt"]

	merge_df =  merge_df.sort_values(by="money_avg", axis=0, ascending=False).reset_index(drop=True)
	merge_df["money_avg_rank"] = merge_df.index
	merge_df = merge_df.sort_values(by="trade_avg", axis=0, ascending=False).reset_index(drop=True)
	merge_df["trade_avg_rank"] = merge_df.index
	
	shop_num = merge_df.shape[0]
	merge_df['label'] = ((merge_df["money_avg_rank"] < shop_num*0.5).apply(lambda x:int(x))) * 10 + (merge_df["trade_avg_rank"] < shop_num*0.5).apply(lambda x:int(x))
	return merge_df
	    
```

```python
def sort_partition(row_df, cluster_dic):
	'''
	cluster_dic, key is the number of partition and value is the set of cardaccptrids and userids.
	according to the RFM model, give each partition a risk label
	'''
	df = row_df.copy()
	print "sort..."
	kpi_data = list()
	for key, value in cluster_dic.items():
		shops = set()
		userids = set()
		for item in value:
			if item.find("shop") != -1:
				shops.add(item.split("_")[1])
			elif item.find("user") != -1:
				userids.add(item.split("_")[1])
			else:
				print key, value
				raise Exception(" 'user' or 'shopt' don't exist... pls set them")
		sub_df = df[(df["cardaccptrid"].isin(shops)) & (df["userid"].isin(userids))]		
		
		sub_trade_num = sub_df.shape[0]
		sub_trade_money = sub_df.outmoney.sum()
		sub_userid_cnt = sub_df.userid.unique().size
		sub_shop_cnt = sub_df.cardaccptrid.unique().size
		try:
			kpi_data.append([key, float(sub_trade_num)/sub_shop_cnt, float(sub_trade_money)/sub_shop_cnt, float(sub_userid_cnt)/sub_shop_cnt])
		except:
			print "shop doesn't exist 。。。"
			print value 
			print 80*"="

	column_names = ["com_num", "num_avg", "money_avg", "cnt_avg"]
	new_df = pd.DataFrame(kpi_data, columns = column_names)
	new_df = new_df.sort_values(by="money_avg", axis=0, ascending=False).reset_index(drop=True)
	new_df["money_avg_rank"] = new_df.index
    
	new_df = new_df.sort_values(by="num_avg", axis=0, ascending=False).reset_index(drop=True)
	new_df["num_avg_rank"] = new_df.index
    
	new_df = new_df.sort_values(by="cnt_avg", axis=0, ascending=True).reset_index(drop=True)
	new_df["cnt_avg_rank"] = new_df.index
	    
	total_cnt = new_df.shape[0]
	new_df['label'] = ((new_df["money_avg_rank"] < (total_cnt * 0.1)).apply(lambda x:int(x)) * 10 + (new_df["num_avg_rank"] < (total_cnt * 0.1)).apply(lambda x:int(x))) * 10 + (new_df["cnt_avg_rank"] < (total_cnt * 0.1)).apply(lambda x:int(x))
	return new_df
```
